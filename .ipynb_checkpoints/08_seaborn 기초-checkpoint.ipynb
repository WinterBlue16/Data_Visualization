# Seaborn 기초
> 데이터 시각화를 위한 대표적 라이브러리인 seaborn과 matplotlib 기초를 학습한다. 

```python
# 라이브러리 불러오기
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# tips 데이터셋 불러오기(sns에서 기본 제공)
tips=sns.load_dataset('tips')
tips.head(6)
```

```python
# 다른 데이터셋도 있음
iris=sns.load_dataset('iris')
titanic=sns.load_dataset('titanic')
flights=sns.load_dataset('flights')
```

```python
# 예시
iris.head(6)
```

## 1. Regplot
- 데이터를 점과 선으로 나타내며 분포와 증감추이를 한번에 확인할 수 있다. 

```python
# 지불 금액에 따른 팁의 양 : ax 객체로 선언 
ax=sns.regplot(x='total_bill', y='tip', data=tips)
# fit_reg=False 선을 출력하지 않는 옵션
# ax=sns.regplot(x="total_bill", y='tip', data=tips, fit_reg=False)
ax.set_xlabel('TB') # x축 이름 설정
ax.set_ylabel('Tip') # y축 이름 설정
ax.set_title('Total bill and Tip')
```

```python
# 선 출력 없음
ax=sns.regplot(x="total_bill", y='tip', data=tips, fit_reg=False)
ax.set_xlabel('TB') # x축 이름 설정
ax.set_ylabel('Tip') # y축 이름 설정
ax.set_title('Total bill and Tip')
```

## 2. Joinplot
- 그래프 안쪽 : 데이터를 점으로 나타내어 분포를 확인할 수 있다.
- 그래프 바깥쪽 : 데이터를 히스토그램으로 나타내어 데이터의 밀집도를 확인할 수 있다. 

```python
joint=sns.jointplot(x='total_bill', y='tip', data=tips)
# kind='hex' 옵션을 주면 점이 아닌 6각 분포로 출력 가능

joint.set_axis_labels(xlabel='TB', ylabel='Tip')
# jointplot에서는 regplot과 라벨 붙이는 방식이 다르다.
# set_axis_labels라는 함수를 사용해 라벨 설정
# 굳이 라벨을 설정하지 않아도 기본 column 명으로 라벨 생성됨
```

```python
# 6각 그래프
joint=sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='TB', ylabel='Tip')
```

## 3. kde : 이차원 밀집도 그래프 
- 등고선 형태로 데이터의 밀집 정도를 확인할 수 있다. 

```python
kde, ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
              data2=tips['tip'],
              shade=True)
ax.set_title('Kernel Density Plot')
```

```python
# shade=False
kde, ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
              data2=tips['tip'],
              shade=False)
ax.set_title('Kernel Density Plot')
```

## 4. Barplot : 막대 그래프 

```python
ax=plt.plot()
ax=sns.barplot(x='time', y='total_bill', data=tips)
```

## Boxplot : 박스 그래프

```python
sns.boxplot(x='day', y='total_bill', data=tips,
           hue='smoker', palette='Set1')
# hue = 'smoker'
# hue 옵션을 사용해서 카테고리별 비교가 가능

# palette='Set3'
# palette 색상 옵션
```

## Pairplot(data=tips)
- 수치에 해당하는 그래프를 전반적으로 그려줌
- column별 관계 그래프를 확인할 수 있음
- 전반적인 상태를 확인할 수 있어 처음 데이터를 분석할 때 전체를 파악하기 좋음

```python
sns.pairplot(data=tips)
```

원하는 그래프와 데이터를 선택해 pairplot을 그리는 것도 가능하다.

```python
pg = sns.PairGrid(tips) # pairgrid 형태 만들기
pg.map_upper(sns.regplot) # 위쪽 그래프에 넣을 plot 
pg.map_lower(sns.kdeplot) # 아래쪽 그래프에 넣을 plot 
pg.map_diag(sns.distplot) # 가운데 그래프에 넣을 plot 
```

## countplot 
- 해당 카테고리별 데이터의 개수를 보여주는 그래프

```python
# 요일별 팁 받은 횟수 출력
sns.countplot(x='day', data=tips)
```

## heatmap
- 카테고리별 데이터 분류

```python
titanic_size=titanic.pivot_table(index='class', columns='sex', 
                                 aggfunc='size')
# aggfunc='size' 각 데이터의 건수에 대해 출력
# 그냥 pivot table은 평균, 분산 등의 결과가 출력됨
titanic_size
```

```python
sns.heatmap(titanic_size, annot=True, fmt='d', cmap=sns.light_palette('blue'))
# annot=True : 숫자가 출력될 수 있게 함
# fmt='d' : 지수 형태의 숫자가 아닌 지수 형태의 숫자로 변경
# cmap=sns.light_palette('blue') : 색상 결정
```

```python
flights.head()
fp=flights.pivot('month', 'year', 'passengers') # 열 인덱스, 행 인덱스, 데이터 순서
sns.heatmap(fp, linewidths=1, annot=True, fmt='d')
```

## Pandas에서 바로 plot 그리기

```python
df=pd.DataFrame(np.random.randn(100,3),
               index=pd.date_range('1/28/2020', periods=100),
               columns=['A', 'B', 'C'])
# 랜덤으로 100행 3열의 Dataframe을 생성
# index를 pd.date_range를 통해서 2020.01.28를 기준으로 100일

# 판다스 데이터프레임에서 각 열에 대해서 그래프를 그릴 수 있음
df.plot() # 일변 변화량 또는 변동폭을 그래프로 나타냄
df.cumsum().plot()
# cumsum() : 누적합 : 누적합으로 그래프를 그리면 해당 열의 값이 어떻게 변해가는 지 확인할 수 있음
# 금융, 주식 등에서 수익률을 계산할 때 활용할 수 있음
```
